```python
from datasets import load_dataset
from transformers import BertTokenizer

# Charger le jeu de données IMDB depuis la bibliothèque `datasets`
dataset = load_dataset("imdb")

# Initialiser le tokenizer BERT
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Fonction pour tokenizer une critique de film
def tokenize_function(example):
    return tokenizer(
        example["text"],
        padding="max_length",  # Ajoute du padding pour une longueur constante
        truncation=True,       # Troncature des critiques trop longues
        max_length=128         # Longueur maximale des tokens
    )

# Appliquer la fonction de tokenization sur tout le dataset IMDB
tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Afficher un exemple tokenisé
print(tokenized_dataset["train"][0])
```

```python
def random_sample(dataset, sample_size, proba_zero=0.5):
    # Utiliser seulement le split 'train'
    train_data = dataset['train']

    # Diviser les données en deux classes : classe 0 (négatif) et classe 1 (positif)
    class_0 = train_data.filter(lambda x: x["label"] == 0)
    class_1 = train_data.filter(lambda x: x["label"] == 1)

    # Calculer la taille d'échantillon pour chaque classe
    size_class_0 = int(sample_size * proba_zero)
    size_class_1 = sample_size - size_class_0  # Le reste est pour la classe 1

    # Prendre un échantillon aléatoire de chaque classe
    sampled_class_0 = class_0.shuffle(seed=42).select(range(size_class_0))
    sampled_class_1 = class_1.shuffle(seed=42).select(range(size_class_1))

    # Combiner les échantillons pour obtenir l'échantillon final
    combined_sample = Dataset.from_dict({
        "text": sampled_class_0["text"] + sampled_class_1["text"],
        "label": sampled_class_0["label"] + sampled_class_1["label"]
    })

    return combined_sample
```

```python
# Créer un échantillon aléatoire de 1000 échantillons équilibrés
sampled_dataset = random_sample(tokenized_dataset, sample_size=1000)

# Afficher les premiers exemples pour vérifier
print(sampled_dataset[:5])
```

```python
import torch
from transformers import BertModel

# Charger le modèle BERT pré-entraîné
model = BertModel.from_pretrained("bert-base-uncased")

def get_bert_embeddings(tokenized_data):
    embeddings = []
    for example in tokenized_data['text']:
        # Tokenisation des critiques
        inputs = tokenizer(example, return_tensors="pt", padding=True, truncation=True, max_length=128)
        with torch.no_grad():
            # Obtenir les embeddings de BERT
            outputs = model(**inputs)
            # On prend le vecteur du token [CLS] comme embedding
            cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze().cpu().numpy()
            embeddings.append(cls_embedding)
    return embeddings

# Obtenir les embeddings pour l'échantillon échantillonné
sampled_embeddings = get_bert_embeddings(sampled_dataset)
```

```python
import torch.nn as nn

# Définition d'un autoencodeur simple
class Autoencoder(nn.Module):
    def __init__(self, input_dim, latent_dim):
        super(Autoencoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.ReLU(True),
            nn.Linear(512, latent_dim)
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 512),
            nn.ReLU(True),
            nn.Linear(512, input_dim),
            nn.Sigmoid()  # Pour une reconstruction dans l'espace d'origine
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Initialisation du modèle d'autoencodeur
latent_dim = 64  # Taille de la dimension latente
input_dim = len(sampled_embeddings[0])  # Taille des embeddings BERT (768 pour bert-base-uncased)
autoencoder = Autoencoder(input_dim, latent_dim)
```

```python
# Conversion des embeddings en tensor PyTorch
embeddings_tensor = torch.tensor(sampled_embeddings)

# Passer les embeddings dans l'autoencodeur pour réduire la dimension
with torch.no_grad():
    reduced_embeddings = autoencoder.encoder(embeddings_tensor).cpu().numpy()

# Afficher quelques embeddings réduits
print(reduced_embeddings[:1])  # Affiche les 5 premiers embeddings réduits
```

```python
import alibi_detect
from alibi_detect.cd import KSDrift

# Configurer le détecteur de dérive KSDrift avec les embeddings réduits
detector = KSDrift(x_ref=reduced_embeddings)  # X_ref contient les embeddings de référence (par exemple, échantillon équilibré)

# Tester la dérive sur un autre jeu de données, par exemple un jeu de données test
detector_result = detector.predict(x=reduced_embeddings, drift_type='batch')  # X_test contient les nouveaux embeddings
print(detector_result)
```
